```python
import pandas as pd
df = pd.read_excel (r'desktop/data.xlsx')
Playerx=[]
for x in range(92):
    Playerx.append((df.Player[x].partition('\\'))[0])
df['NPlayer']=Playerx
df.drop('Player', inplace=True, axis=1)
df=df[['NPlayer','Pos', 'npxG', 'xA', 'npxG+xA', 'xA p90', 'npxG p90', 'npxG+xA p90']]
df.rename(columns = {'NPlayer':'Player'}, inplace = True)
df.to_excel("Cleaned_Data.xlsx")
```
